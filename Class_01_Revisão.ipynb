## 1. Introdução -  Objetivo da disciplina <a name="section_1"></a>


- Nessa disciplina procura-se fornecer uma base teórica adequada para a compreensão dos fenômenos da área de análise estrutural que estão inter-relacionados com as áreas de Engenharia Civil, Mecânica, de Minas, etc... visando fixar conceitos básicos que possibilitem a troca de informações entre profissionais de áreas diferentes.

- Outros: projetos de embalagem, logística de transporte e de manuseio de carga, problemas estruturais nas torres de transmissão e nos cabos de linha de transmissão, instrumentação de ensaios estruturais, bases de motores, postes de transformadores.

A Mecânica Estrutural permite estudar a resistência, a rigidez e a estabilidade das peças e das estruturas para obter o projeto mais seguro e mais econômico.

Ela se baseia nas leis de equilíbrio e nos resultados da mecânica dos sólidos deformáveis. A mecânica dos sólidas permite tratar de forma rigorosa a maioria dos problemas mas o o quadro matemático necessário é em geral complexo dificultando os cálculos, exigindo o uso de ferramentas computacionais. 

A Mecânica Estrutural propõe métodos de cálculo mais simples e mais rápidos que conduzem a resultados aproximados mas adaptados à precisão exigida na prática. 

O Curso é organizado da seguinte forma:

- Análise dos esforços internos nas estruturas estaticamente determinadas
- Análise de tensões e deformações nas estruturas estaticamente determinadas
- Análise de sistemas complexos de tensões e deformações produzidos por carregamentos de tipos diferentes aplicados simultaneamente.

O conhecimento adquirido nesta disciplina é fundamental para uma boa compreensão dos resultados obtidos em análise realizada com programas computacionais. É importante para um bom desempenho em disciplinas como Concreto Armado e Protendido, Aço e Maderia, Alvenaria Estrutural, entre outras.

As hipóteses admitidas em Mecânica Estrutural se baseiam na experiência ou se justificam com teorias matemáticas. São hipóteses gerais relativas aos materiais utilizados, à forma dos sólidos e aos tipos de ações mecânicas aplicadas.

#### Hipóteses sobre os materiais:

Continuidade do material: o material é contínuo, homogêneo e isótropo

Características mecânicas dos materiais: são determinadas experimentalmente com ensaios normalizados que permitem qualificar e mensurar os parâmetros de comportamento. Por exemplo: o ensaio de tração ou de compressão simples;

#### Hipóteses sobre as ações externas:

Toda ação mecânica é representada por um tensor em um ponto. As ações podem ser concentradas ou distribuídas, exercidas à distância ou em contato.

A figura abaixo ilustra uma edificação básica onde são identificados diferentes tipos de elementos que serão definidos na sequência:

<img src="images/elementos.png" alt="elementos" width="553px"/>

**_Barras_**: duas dimensões pequenas com relação à terceira; podem ser consideradas unidimensionais, e representadas pelo seu eixo (lugar geométrico dos centros de gravidade de suas seções transversais).  A dimensão maior (longitudinal) é o comprimento. As duas dimensões menores definem a seção transversal. Exemplos de barras: vigas, pilares, arcos, tirantes, escoras.

**_Placas e cascas_**: uma dimensão pequena com relação às outras duas; podem ser consideradas bi-dimensionais, sendo representadas por sua superfície média. Exemplos de placas (superfície plana): lajes, paredes.  A dimensão menor é a espessura. As outras duas definem a superfície média. Exemplos de cascas (superfícies curvas): coberturas, silos, reservatórios cilíndricos.

**_Blocos_**: três dimensões equivalentes; são elementos tridimensionais. Exemplos: blocos de fundações, bases de equipamentos, barragens.

Uma estrutura pode ser constituída por um único tipo de elementos ou por uma combinação de vários tipos de elementos. Nesta disciplina vamos estudar apenas as barras.

#### Princípio de equivalência da ação e reação

As forças podem ser classificadas em _externas_ e _internas_:
- As _forças externas_ representam a ção de outros corpos sobre o objeto considerado, e são responsáveis pelo comportamento do corpo. Elas podem ser _ativas_ ou _reativas_. As _forças ativas_ são aquelas aplicadas diretamente sobre o objeto considerado e as _forças reativas_ são aquelas que se originam por influência de vínculos e que só aparecem quando atuam as forças ativas. Desta forma, qualquer modificação nas forças ativas provoca uma modificação nas  forças reativas .
- As _forças internas_ são as que mantém unidas as partes que formam o corpo considerado.

## 5. Equilíbrio estático <a name="section_5"></a>

#### Condições de equilíbrio  
Para que um corpo submetido a um sistema de forças esteja em equilíbrio é necessário que este conjunto de forças não provoque nenhuma tendência de translação nem de rotação.

<img src="images/corpo livre.png" alt="187px" width="336px"/>

Como a tendência de translação é dada pela força resultante R e a tendência de rotação em torno de qualquer ponto é dada pelo momento resultante M destas forças em relação ao ponto considerado, basta que estes vetores R e M sejam nulos para que o corpo esteja em equilíbrio.

Assim, a condição necessária e suficiente para que um corpo esteja em equilíbrio, quando submetida a ação de um sistema de forças é que  estas forças satisfazem as seguintes equações :  


$ \left\{ \begin{align*}  
 \sum\,F_x     &= 0   \\
 \sum\,F_y     &= 0   \\
 \sum\,F_z     &= 0   
\end{align*} \right. \quad \mbox{e}\quad$     $\left\{ \begin{align*}   
 \sum\,M_x     &= 0   \\
 \sum\,M_y     &= 0   \\
 \sum\,M_z     &= 0   
\end{align*} \right.  $

#### Graus de liberdade  
Conforme foi visto, no caso tridimensional tem-se 6 equações de equilíbrio que estão relacionadas com as três componentes de forças e as três componentes de momentos, que por sua vez estão relacionados com as três translações possíveis e as três rotações possíveis. Isto define os 6 graus de liberdade do corpo no espaço, que são os 6 movimentos possíveis que ele pode ter. No plano temos só três graus.
<img src="images/graus de liberdade.png" alt="220px" width="745px"/>

Para que o corpo esteja em equilíbrio (ausência de movimento), estes 6 graus de liberdade devem ser restringidos pela ação dos apoios os quais devem ser capazes de impedir todas as tendências de movimento. O efeito dos apoios é o de fazer surgir as reações sobre o corpo, que se oporão às cargas aplicadas formando um conjunto de forças (cargas e reações) em equilíbrio ($R =0$ e $M =0$). 

Para o caso bidimensional, existem 3 graus de liberdade: 2 de translações nas direções dos eixos x e y e 1 de rotação no plano xy em torno do eixo z.

## 6. Vinculação <a name="section_6"></a>
#### Conceito de Vínculos

Vínculo é todo elemento de ligação entre as partes componentes de uma estrutura ou entre a própria estrutura e o meio exterior. A função dos vínculos é a de restringir os graus de liberdades para o qual eles fazem surgir reações nas direções dos movimentos impedidos. Toda vez que há qualquer restrição ao movimento há um vínculo. Os vínculos causam as forças reativas ou reações, sendo claro que só há reação quando há movimento impedido. os vínculos então limitam a mobilidade da estrutura e apresentam reações nas direções dos movimentos que impedem. Os vínculos tem como finalidade:
* transmitir forças (estática);  
* estabelecer uma ligação (física).

Conforme a posição em que os vínculos aparecem em uma estrutura eles são denominados de ligações (quando unem dois elementos da mesma estrutura) ou apoios (quando unem a própria estrutura com o meio externo).

#### Classificação dos vínculos

Como foi visto, a função dos vínculos é restringir alguns graus de liberdade da estrutura, fazendo surgir reações segundo as direções dos movimentos impedidos. Os vínculos são classificados segundo os movimentos que impedem. Os movimentos impedidos podem ser de translação ou de rotação e eles definem o número de reações que aparecem e a  classe do vínculo.

Número de movimentos impedidos ==> Número de reações ==> Classe do vínculo

Como nosso estudo ficará limitado ao plano, os vínculos que nos interessam são os de 1$^a$, 2$^{da}$ e 3$^{ra}$ classe (1, 2 e 3 movimentos restringidos respectivamente).

**_Vínculo de 1a classe_**: restringe 1 movimento, faz surgir 1 reação, tem dois graus de liberdade. É denominado de apoio simples: pode impedir um translação, permitindo uma translação e uma rotação. Também constituem vínculos de 1$^a$ classe, ou podem ser considerados como tais, estruturalmente, os seguintes casos: hastes e tirantes que impedem a translação na direção de seu eixo. A figura mostra exemplos e representação de vínculos de 1$^a$ classe.

<img src="images/vinculo classe 1.png" alt="300px" width="600px"/>

**_Vínculo de 2$^{da}$ classe_**: restringe 2 movimentos, introduz 2 reações, tem 1 grau de liberdade; é denominado de articulação ou rótula: pode impedir duas translações, permitindo uma rotação. Também constituem vínculos de 2$^{da}$ classe dois tirantes cujas prolongações se intersectam em um ponto, em torno do qual a a estrutura pode girar.

<img src="images/vinculo classe 2.png" alt="300px" width="600px"/>

**_Vínculo de 3$^{ra}$ classe_**: restringe os 3 movimentos , introduz três reações, não tem grau de liberdade ( grau de liberdade = 0); é denominado de engastamento ou engaste: impede translações e rotação. Também constituem vínculos de 3$^{ra}$ classe três tirantes ligados a mesma estrutura.

<img src="images/vinculo classe 3.png" alt="300px" width="600px"/>

**_Vínculos perfeitos e imperfeitos_**: na prática é muito difícil ter um vínculo perfeito. O vínculo é denominado imperfeito quando ele permite movimentos que deveria restringir (ainda que sejam pequenos) ou restringe movimentos que deveria permitir. é o caso de recalque de apoio. No caso de uma ponte ferroviária: nos nós, admitem-se ligações de 2$^a$ classe, porém se utilizam rebites ou parafusos em vez de pinos, os quais constituem ligações de 3$^a$ classe.

## 7. Estaticidade e estabilidade  <a name="section_7"></a>

Conforme foi visto, a função dos apoios é restringir os graus de liberdade de uma estrutura. Assim, três casos podem ocorrer:

1. _Os apoios estão em número estritamente necessário_ para impedir todos os movimentos possíveis da estrutura. Neste caso, o número de reações de apoio a determinar (incógnitas) é igual ao número de equações de equilíbrio disponíveis.  O  sistema de equações é determinado. A estrutura se denomina isostática e o equilíbrio é estável [esquema (a) da figura].  
  
2. _Os apoios estão em número superior ao necessário_ para impedir todos os movimentos possíveis da estrutura. Neste caso, o número de reações de apoio a determinar (incógnitas) é superior ao número de equações de equilíbrio disponíveis. O sistema de equações indeterminado. A estrutura, neste caso, se denomina hiperestática e o equilíbrio é estável [esquema (b) da figura]. Como as equações de equilíbrio não são suficientes para determinar as reações, é necessário utilizar equações adicionais (equações de compatibilidade de deformações).
3. _Os apoios estão em número inferior ao necessário_ para impedir todos os movimentos possíveis da estrutura. Neste caso, o número de reações de apoio a determinar (incógnitas) é inferior ao número de equações de equilíbrio disponíveis configurando um sistema de equações impossível (sem solução). A estrutura neste caso se denomina  hipoestática e o equilíbrio (se existir) é instável [esquema (c) da figura].

<img src="images/estaticidade.png" alt="255px" width="740px"/>


É importante ressaltar que não basta uma quantidade adequada de vínculos, mas também a sua correta distribuição. Normalmente, o grau de estaticidade g de uma estrutura é determinado como o somatório dos graus externo ge e interno gi. Detalhes sobre a determinação dos graus interno e externo podedem ser encontrados no livro de Masuero e Creus, pagina 29, o qual está disponível em pdf no moodle.

No âmbito da disciplina, onde o interesse é o equilíbrio estático, o grau de estaticidade pode ser determinado de uma forma simples e rápida considerando que qualquer estrutura é constituida por uma ou várias partes unidas por ligações. A figura mostra exemplos de ligações com as restrições que introduzem.

<img src="images/ligações.png" alt="257px" width="504px"/>

Cada corpo permite escrever 3 equações de equilíbrio e tem-se portanto  $3\cdot n$ equações. Considerando $r$ (reações de apoios) e $v$ (reações entre partes) a estrutura é determinada se:  $r + v =3n$. Desta forma, fazendo $g=3n - (r+v)$, teremos:

* $g = 0$  Estrutura isostática  


* $g > 0$  Estrutura hipoestática 


* $g < 0$  Estrutura hiperestática  


## 8. Reações de apoio <a name="section_8"></a>
#### Métodos de resolução

Para a resolução de um problema de resistência dos materiais, é necessário primeiro resolver o problema de equilíbrio estático e depois um problema de dimensionamento. Portanto deve-se calcular primeiro as reações necessárias para manter a estrutura em equilíbrio. Para isso é necessário identificar todas as forças aplicadas (ativas)  e substituir os vínculos por forças reativas correspondentes (incógnitas). O esquema estrutural assim definido com todas as forças ativas e reativas é denominado de Diagrama de corpo livre (DCL). 

<img src="images/dcl.png" alt="257px" width="504px"/>

Na sequência aplicam-se as equações de equilibrio que permitem determinar os valores das reações para garantir oequilíbrio, as quais são:


$ \left\{ \begin{align*}  
 \sum\,F_x     &= 0   \\
 \sum\,F_y     &= 0   \\
 \sum\,F_z     &= 0   
\end{align*} \right. \qquad \mbox{e}\qquad$     $\left\{ \begin{align*}   
 \sum\,M_x     &= 0   \\
 \sum\,M_y     &= 0   \\
 \sum\,M_z     &= 0   
\end{align*} \right.  $




## 9. Carregamentos <a name="section_9"></a> 

Para a resolução de um problema de resistência dos materiais, é necessário primeiro resolver o problema de equilíbrio estático e depois um problema de dimensionamento. Portanto deve-se calcular primeiro as reações necessárias para manter a estrutura em equilíbrio. Para isso é necessário identificar todas as forças aplicadas e substituir os vínculos por forças reativas correspondentes. A seguir algumas considerações sobre os carregamentos.

### 9.1 Carga concentrada <a name="section_91"></a>
São forças que atuam em pontos determinados e de forma constante. Na prática uma carga sempre se distribui sobre um trecho de estrutura mas se este trecho for desprezível em relação à dimensão da estrutura, a carga pode ser considerada concentrada. Exemplo: um caminhão carregado transmitirá a carga mais seu peso próprio através das rodas cujo contato com o solo ocorre sobre um comprimento muito pequeno em relaçã ao comprimento do caminhão. Neste caso, as forças nas rodas são cargas concentradas.

<img src="images/caminhão-conc.png" alt="150px" width="350px"/>

Unidades de uma carga concentrada: kgf, N, kN, tf ; kgf = 9,81 N

### 9.2 Carga distribuida com linha de carga qualquer <a name="section_92"></a>  
Se o trecho de distribuição da carga não for desprezível em relação à dimensão total da estrutura, ela pode ser considerada distribuída,  como mostra a figura onde cada livro tem uma altura diferente, a qual pode ser considerada como a carga na posição.

<img src="images/carga distribuida qual.png" alt="260px" width="430px"/>

O esquema abaixo mostra uma distribuição parabólica onde $dF$ é a carga que atua no trecho infinitessimal $dx$.  

<img src="images/carga distr parab.png" alt="250px" width="250px"/>

Uma carga distribuída é caracterizada pela taxa de distribuição $q$, de tal forma que:

$$dF=q\cdot dx \qquad \Rightarrow \qquad q=\frac{dF}{dx}$$

Unidades para uma carga distribuida linear: kgf/m, N/m, kN/m, tf/m

#### Determinante da resultante de carregamentos distribuídos

Uma carga distribuída pode ser reduzida a uma resultante única cujo módulo é o somatório das forças que a compõem.

$$ R=\int_{a}^{b} q\cdot dx$$

O valor numérico da resultante é a área da superfície delimitada pela linha de carga e pelo eixo da estrutura. Esta superfície é denominada superfície de carga. O ponto de aplicação da resultante pode ser determinado aplicando-se o teorema de Varignon que estabelece que o momento da resultante de um sistema de força é igual à soma dos momentos de cada uma das forças:

$$ R\cdot d=\sum q\,dx\cdot x $$

$$ \int_{a}^{b} q\cdot dx\cdot d= \int_{a}^{b} q\,x\, dx$$

$$d=\frac{\displaystyle{\int_{a}^{b} q\,x\, dx}}{\displaystyle{\int_{a}^{b} q\,dx}}$$

Assim, a resultante de uma carga distribuída é igual à área da superfície de carga e passa pelo centro de gravidade desta superfície.

### 9.3 Carga distribuida uniforme <a name="section_93"></a> 
A carga uniformemente distribuída quando sua taxa de distribuição  é constante ($q= cte$). A  linha de carga é paralela ao eixo da estrutura. São exemplos de cargas distribuidas uniformes; a carga de uma parede apoiada sobre uma viga, o peso próprio de uma viga de seção constante. Neste caso a carga linear se calcula fazendo $q = seção\, x\, peso\, especifico$.

<img src="images/parede alv.png" alt="133px" width="340px"/>

Também é possível transformar uma carga distribuida com linha de carga qualquer em uma carga distribuida uniformemente por trecho. O exemplo do estante de livro teria dois trechos de carga distribuida uniforme.

<img src="images/carga distribuida.png" alt="250px" width="500px"/>

Para uma carga uniforme distribuída sobre um comprimento $\ell$ como mostra o esquema:

<img src="images/repr carga distr uni.png" alt="180px" width="265px"/>

A resultante é dada por $R=q\cdot \ell$

e a posição é determinada considerando :


$$R\cdot d=\sum q\,dx\cdot x $$

$$ \int_{0}^{\ell} q\cdot dx\cdot d= \int_{0}^{\ell} q\,x\, dx$$

$$d=\frac{\displaystyle{\int_{0}^{\ell} q\,x\, dx}}{\displaystyle{\int_{0}^{\ell} q\,dx}}=\frac{1}{2}\cdot \ell$$






### 9.4 Carga distribuída triangular <a name="section_94"></a>  

Tem-se uma carga distribuída triangular quando a taxa de distribuição da carga é uma função linear de $x$ ($q=f(x)$). A linha de carga forma um triângulo com o eixo da estrutura.

<img src="images/carga distribuida tri.png" alt="160px" width="580px"/>

São exemplos de carga distribuida triangular: a carga hidrostática sobre as paredes de reservatórios, a carga de solos em muros de contenção.

No caso de uma carga distribuida triangular sobre um comprimento $\ell$ (No esquema acima $\ell$ teria o valor de 3 m), a resultante vale:
$$R=\frac{q_t\cdot\ell}{2}$$
e a posição desta resultante resulta da seguinte análise. É necessário estabelecer o valor da carga distribuida em um ponto qualquer de posição x:

$$\frac{q_x}{x}=\frac{q_t}{\ell}\quad \Rightarrow \quad q_x=\frac{q_t}{\ell}\,x$$

Aplicando o teorema de Varignon:

$$R\cdot d=\sum \,q_x\, dx\cdot x $$

$$\frac{q_t\,\ell}{2}\cdot d=\int_0^{\ell}\frac{q_t\cdot x}{\ell}\cdot x\;dx =\frac{q_t}{\ell}\,\int_0^{\ell} x^2\;dx$$

### 9.6 Superposição de efeitos <a name="section_95"></a>

O princípio da superposição de efeitos depende do princípio da independência da ação das forças segundo a qual cada força age independentemente da ação das demais forças. Este estabelece que o efeito produzido por todas as cargas que agem sobre uma estrutura é igual à soma dos efeitos de cada uma das forças agindo isoladamente. O princípio é aplicável para o caso de pequenas deslocamentos($< L/400$) e com a estrutura no regime elástico. A figura mostra a transformação de do carregamento de uma viga em um sistema equivalente evidenciando a superposição de efeito.
